```python
df.drop(['CUST_ID','MINIMUM_PAYMENTS'],axis=1,inplace=True)
df.head(3)
```

```python
df['CREDIT_LIMIT']=df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mode()[0])
df.isna().sum()
```

```python
x=df
x
```

```python
from sklearn.cluster import KMeans
#create an empty list
wcs=[]
for i in range(1,11):
  model =KMeans(n_clusters=i,init='k-means++',random_state=42)
  model.fit(x)
  wcs.append(model.inertia_) #model.inertia_ used to get the variance of the model
```

```python
wcs# all variance corresponding to the cluster numbers from 1 to 10
```

```python
plt.plot(range(1,11),wcs,'-*y',mfc='b')
plt.xlabel('Number of Clusters', color='m')
plt.ylabel('variance',color='g')
plt.title('Elbow graph')
```

```python
model1=KMeans(n_clusters=9,init='k-means++',random_state=42)
model1.fit(x)
y_pred=model1.predict(x)
y_pred
```

```python
data=x
data
```

```python
data['Output']=y_pred
data
```

```python
data.isna().sum()
```

```python
y=data['Output']
y.ndim
```

```python
x=data.drop(['Output'],axis=1)
print(x.ndim)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
mod1=SVC()
mod1.fit(x_train,y_train)
y1_pred=mod1.predict(x_test)
y1_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y1_pred)*100)
```
